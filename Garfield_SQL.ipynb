### Section 2: WHERE

```python
# Section 2: WHERE. Complete 17.2.3 WHERE (3 queries)

pd.read_sql("""SELECT title, edition, copyright 
                FROM titles
                WHERE copyright > '2016'""", connection)
```

```python
pd.read_sql("""SELECT id, first, last
                FROM authors
                WHERE last LIKE 'D%'""",
                connection, index_col=['id'])
```

```python
pd.read_sql("""SELECT id, first, last
                FROM authors
                WHERE first LIKE '_b%'""",
                connection, index_col=['id'])
```

### Section 3: ORDER BY

```python
# Section 3: ORDER BY. Complete 17.2.4 ORDER BY (4 queries)

pd.read_sql('SELECT title FROM titles ORDER BY title ASC',
            connection)
```

```python
pd.read_sql("""SELECT id, first, last
                FROM authors
                ORDER BY last, first""",
                connection, index_col=['id'])
```

```python
pd.read_sql("""SELECT id, first, last
                FROM authors
                ORDER BY last DESC, first ASC""",
                connection, index_col=['id'])
```

```python
pd.read_sql("""SELECT isbn, title, edition, copyright
                FROM titles
                WHERE title LIKE '%How to Program'
                ORDER BY title""", connection)
```

### Section 4: INNER JOIN

```python
# Section 4: INNER JOIN. Complete 17.2.5 INNER JOIN (1 query)

pd.read_sql("""SELECT first, last, isbn
                FROM authors
                INNER JOIN author_ISBN
                    ON authors.id = author_ISBN.id
                ORDER BY last, first""", connection).head()
```

### Section 5 INSERT INTO

```python
# Section 5: INSERT INTO. Complete 17.2.6 INSERT INTO (2 queries)

cursor = connection.cursor()

cursor = cursor.execute("""INSERT INTO authors (first, last)
                            VALUES ('Sue', 'Red')""")
```

### Section 6: UPDATE

```python
# Section 6: UPDATE. Complete 17.2.7 UPDATE (2 queries)

cursor = cursor.execute("""UPDATE authors SET last='Black'
                        WHERE last='Red' AND first='Sue'""")

cursor.rowcount
```
